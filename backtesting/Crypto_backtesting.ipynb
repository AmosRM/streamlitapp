```python
import pandas as pd
from binance import Client
from sqlalchemy import create_engine
```

```python
client = Client()
```

```python
engine = create_engine('sqlite:///Crypto_1.db')
```

```python
def getminutedata(symbol,start):
    frame = pd.DataFrame(client.get_historical_klines(symbol,'1m', start))
    
    frame = frame[[0,1,2,3,4]]
    frame.columns = ['Date','Open','High','Low','Close']
    frame.Date = pd.to_datetime(frame.Date, unit='ms')
    frame.set_index('Date', inplace=True)
    frame = frame.astype(float)
    return frame
```

```python
df = getminutedata('BTCUSDT','1 days ago UTC')
```

```python
df
```

```python
df.resample('2min').agg({
    'Open':'first',
    'High':'max',
    'Low':'min',
    'Close':'last'
})
```
