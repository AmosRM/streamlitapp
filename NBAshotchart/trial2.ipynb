```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

from shot_chart import draw_court, get_league_shots, shot_chart
```

```python
df = pd.read_csv('shots_21.csv')
```

```python
threshold = 100
shot_chart_options = {
    'threshold': threshold,
    'color': 'pct',
    'markersize': [25, 100],
    'nbins': 4,
    'gridsize': 50,
    'legend_args': {'columnspacing': 0.1},}
```

```python
season = '2020-21'
```

```python
title = f'NBA Leage-wide Shot Chart: {season}\nMinimum {threshold} Attempts'

fig, ax = plt.subplots(figsize=(12, 8), dpi=100)
shot_chart(df, ax, title=title, **shot_chart_options)
draw_court(ax=ax)
plt.xlim(-260, 260)
plt.ylim(-60, 300)
plt.axis('off')
plt.show()
```
