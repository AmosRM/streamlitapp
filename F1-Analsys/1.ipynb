```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import fastf1 as ff
from fastf1 import plotting
```

```python
ss = ff.get_session(2022,'Dutch Grand Prix','R')
```

```python
dutch_race = ss.load_laps()
```

```python
HAM = dutch_race.pick_driver('HAM')
VER = dutch_race.pick_driver('VER')
```

```python
VER
```

```python
VER = VER[VER['LapNumber'] > 1]
HAM = HAM[HAM['LapNumber'] > 1]
```

```python
plotting.setup_mpl()
```

```python
plt.rcParams['figure.figsize'] = [15, 15]

fig, ax = plt.subplots()

ax.plot(HAM['LapNumber'], HAM['LapTime'], label='HAM', color='cyan')
ax.plot(VER['LapNumber'], VER['LapTime'], label='VER', color='red')
ax.legend()

plt.show()
```
